```python
#Import necessary libraries
import os
from google.colab import userdata
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

#Set OpenAI API key
os.environ['GEMINI_API_KEY'] = userdata.get('GEMINI_API_KEY')

#Initialize the ChatGeminiAI model
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)
```

```python
from langchain_core.prompts import ChatPromptTemplate

#Create a prompt template
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant."),
        ("user", "{question}")
    ]
)
```

```python
from langchain_core.output_parsers import StrOutputParser

# Initialize the string output parse
parser = StrOutputParser()
```

```python
# Chain the prompt, LLM, and output parser

chain = prompt | llm | parser
```

```python
question = "What is n8n?"

response = chain.invoke({"question": question})

print(response)
```

```python
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

# Create a prompt template using MessagePlaceHolder for the question
prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="Your are an intelligent chatbot"),
        MessagesPlaceholder(variable_name="question"),
    ]
)
```

```python
# Chain the prompt, LLM, and output parser
chain = prompt | llm | parser
```

```python
question = "My name is KC"

response = chain.invoke({"question": [HumanMessage(content=question)]})

print(response)
```

```python
question = "who am I"

response = chain.invoke({"question": [HumanMessage(content=question)]})

print(response)
```

```python
# Create a prompt template with a predefined conversation history and a new question placeholder

prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="Your are an intelligent chatbot"),
        HumanMessage(content="My name is KC"),
        AIMessage(content="Nice to meet you KC"),
        MessagesPlaceholder(variable_name="question")
    ]
)
```

```python
# Chain the prompt, LLM, and output parser
chain = prompt | llm | parser
```

```python
question = "who am I"

response = chain.invoke({"question": [HumanMessage(content=question)]})

print(response)
```

```python
# Create a prompt template with a dynamic conversation history and new question placeholder

prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="Your are an intelligent chatbot"),
        MessagesPlaceholder(variable_name="history"),
        MessagesPlaceholder(variable_name="question")
    ]
)
```

```python
# Chain the prompt, LLM, and output parser
chain = prompt | llm | parser
```

```python
# Define the conversation history
history = [
    HumanMessage(content="My name is KC"),
    AIMessage(content="Nice to meet you KC"),
    HumanMessage(content="What is 8 + 4"),
    AIMessage(content="12"),
]
```

```python
question = "who am I"

response = chain.invoke({"history": history,"question": [HumanMessage(content=question)]})

print(response)
```

```python
question = "what was my last question"

response = chain.invoke({"history": history,"question": [HumanMessage(content=question)]})

history.extend([HumanMessage(content=question), AIMessage(content=response)])

print(response)
```
